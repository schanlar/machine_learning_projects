# Splitting and Scaling the data

```python
from sklearn.model_selection import train_test_split
```

```python
# The features
X = df.drop("benign_0__mal_1", axis=1).values
y = df["benign_0__mal_1"].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=101)
```

```python
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()

X_train = scaler.fit_transform(X_train)

X_test = scaler.transform(X_test)
```

# Creating a Neural Network Model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
```

```python
model = Sequential()

model.add(Dense(X_train.shape[1], activation="relu"))
model.add(Dense(X_train.shape[1]//2, activation="relu"))

# BINARY CLASSIFICATION (MUTUALLY EXCLUSIVE CLASSES)
model.add(Dense(1, activation="sigmoid"))

model.compile(optimizer="adam", loss="binary_crossentropy")
```

```python
model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=600)
```

```python
df_losses = pd.DataFrame(model.history.history)
df_losses.head()
```

```python
df_losses.plot()
```

For this particular data set, training for 600 epochs resulted in overfitting the data. Obviously, we trained too much our model. To remedy this, we are going to use an early stopping condition. This will basically stop the training of our model once it starts getting out of hand.

```python
# Re-define our model
model = Sequential()

model.add(Dense(X_train.shape[1], activation="relu"))
model.add(Dense(X_train.shape[1]//2, activation="relu"))

# BINARY CLASSIFICATION (MUTUALLY EXCLUSIVE CLASSES)
model.add(Dense(1, activation="sigmoid"))

model.compile(optimizer="adam", loss="binary_crossentropy")
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
help(EarlyStopping)
```

```python
# define the condition for early stopping
early_stop = EarlyStopping(monitor="val_loss", mode="min", verbose=1, patience=25)
```

```python
model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=600, callbacks=[early_stop])
```

```python
df_loss = pd.DataFrame(model.history.history)
df_loss.plot()
```

Another thing we could do to prevent overfitting is adding dropout layers. Dropout layers will essentially turn-off a percentage of neurons randomly.

```python
# Re-define our model
model = Sequential()

model.add(Dense(X_train.shape[1], activation="relu"))
model.add(Dropout(0.5))
model.add(Dense(X_train.shape[1], activation="relu"))
model.add(Dropout(0.5))


# BINARY CLASSIFICATION (MUTUALLY EXCLUSIVE CLASSES)
model.add(Dense(1, activation="sigmoid"))

model.compile(optimizer="adam", loss="binary_crossentropy")
```

```python
model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=600, callbacks=[early_stop])
```

```python
df_loss = pd.DataFrame(model.history.history)
df_loss.plot()
```

It is clear that by adding dropout layers and imposing early stopping we have drastically improve the performance of our model.

# Evaluation of model

```python
# For binary classification (i.e. sigmoid last-layer activation function)
y_pred = (model.predict(X_test) > 0.5).astype("int32")

# For multi-class classification (i.e. softmax last-layer activation function)
# y_pred = np.argmax(model.predict(X_test), axis=1)

y_pred
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test, y_pred))
```

```python
print(confusion_matrix(y_test, y_pred))
```

We notice that our model performs extremely well and that it misclassified only 3+1 = 4 cases.
